```python
import os
import requests

image_folder = 'images'
annotation_folder = 'annotations'

# Create the image and annotation folders if they don't exist
os.makedirs(image_folder, exist_ok=True)
os.makedirs(annotation_folder, exist_ok=True)

# Read the document file
with open('birdsnap_content/images.txt', 'r') as file:
    lines = file.readlines()

# Iterate through each line in the document
for i, line in enumerate(lines):
    if i > 0:  # Skip the header line
        fields = line.strip().split('\t')
        image_url = fields[0]
        image_filename = f"{str(i).zfill(5)}.jpg"
        annotation_filename = f"{str(i).zfill(5)}.txt"

        # Download the image and handle 404 errors
        try:
            response = requests.get(image_url)
            response.raise_for_status()
        except requests.exceptions.HTTPError as err:
            print(f"Error downloading image {image_url}: {err}")
            continue

        # Save the image
        image_path = os.path.join(image_folder, image_filename)
        with open(image_path, 'wb') as image_file:
            image_file.write(response.content)

        # Create the annotation file
        annotation_path = os.path.join(annotation_folder, annotation_filename)
        with open(annotation_path, 'w') as annotation_file:
            annotation_file.write(fields[3] + ' ' + fields[4] + ' ' + fields[5] + ' ' + fields[6])

print("Processing completed!")
```

```python
# translate coordinates to YOLO format

import os

annotation_folder = 'annotations'

# Iterate through each txt file in the annotations folder
for filename in os.listdir(annotation_folder):
    if filename.endswith('.txt'):
        annotation_path = os.path.join(annotation_folder, filename)
        
        # Read the contents of the annotation file
        with open(annotation_path, 'r') as file:
            contents = file.read().strip().split(' ')

        # Extract the bounding box coordinates
        bb_x1, bb_y1, bb_x2, bb_y2 = map(int, contents[1:5])

        # Compute the center coordinates, width, and height
        center_x = (bb_x1 + bb_x2) // 2
        center_y = (bb_y1 + bb_y2) // 2
        width = bb_x2 - bb_x1
        height = bb_y2 - bb_y1

        # Update the contents of the annotation file
        contents.extend([str(center_x), str(center_y), str(width), str(height)])
        updated_contents = ' '.join(contents)

        # Write the updated contents back to the annotation file
        with open(annotation_path, 'w') as file:
            file.write(updated_contents)

print("Processing completed!")
```

```python
import os
from PIL import Image

annotation_folder = 'annotations'
image_folder = 'images'

# Iterate through each txt file in the annotations folder
for filename in os.listdir(annotation_folder):
    if filename.endswith('.txt'):
        annotation_path = os.path.join(annotation_folder, filename)
        image_path = os.path.join(image_folder, filename[:-4] + '.jpg')

        # Read the dimensions of the associated image
        with Image.open(image_path) as img:
            image_width, image_height = img.size

        # Read the contents of the annotation file
        with open(annotation_path, 'r') as file:
            contents = file.read().strip().split(' ')

        # Check if the annotation file already has correct values and formatting
        if len(contents) >= 9:
            center_x, center_y, width, height = map(float, contents[-4:])
            if (
                0 <= center_x <= 100 and
                0 <= center_y <= 100 and
                0 <= width <= 100 and
                0 <= height <= 100
            ):
                continue  # Skip modifying the file if it has correct values and formatting

        # Convert the last four fields to percentage values
        center_x_percent = (float(contents[-4]) / image_width) * 100
        center_y_percent = (float(contents[-3]) / image_height) * 100
        width_percent = (float(contents[-2]) / image_width) * 100
        height_percent = (float(contents[-1]) / image_height) * 100

        # Update the contents of the annotation file
        contents[-4:] = [
            str(round(center_x_percent, 2)),
            str(round(center_y_percent, 2)),
            str(round(width_percent, 2)),
            str(round(height_percent, 2))
        ]
        updated_contents = ' '.join(contents)

        # Write the updated contents back to the annotation file
        with open(annotation_path, 'w') as file:
            file.write(updated_contents)

print("Processing completed!")
```

```python
import os
import imagehash
from PIL import Image
import logging

# Configure logging
logging.basicConfig(filename='error.log', level=logging.ERROR)

# Specify the folder containing the images
image_folder = 'images'

# Get a list of image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.endswith('.jpg')]

# Loop through the first 5 images
for i, image_file in enumerate(image_files[:5]):
    image_file_path = os.path.join(image_folder, image_file)

    try:
        # Open the image and calculate the perceptual hash
        img = Image.open(image_file_path)
        phash = imagehash.phash(img)

        # Output the perceptual hash
        print(f"Image {i+1}: {image_file}\tPerceptual Hash (phash): {phash}")

        # Close the image
        img.close()

    except Exception as e:
        # Log any errors that occur
        logging.error(f"Error processing image {image_file}: {str(e)}")

print("Processing completed!")
```

```python
import hashlib

def calculate_md5(file_path):
    with open(file_path, 'rb') as file:
        md5_hash = hashlib.md5()
        while True:
            data = file.read(4096)
            if not data:
                break
            md5_hash.update(data)
    return md5_hash.hexdigest()

# Provide the path to the JPG image file
image_file_path = 'images/00001.jpg'

# Calculate the MD5 hash value
md5_hash_value = calculate_md5(image_file_path)

# Print the MD5 hash value
print("MD5 Hash Value:", md5_hash_value)
```

```python
import os
import shutil

hash_folder = 'images'  # Folder containing the JPG images
second_file_path = 'birdsnap_content/images.txt'  # Path to the second text file

output_folder = 'annotations_hash_output_large'  # Folder to store the new text files
os.makedirs(output_folder, exist_ok=True)

# Loop through the JPG images
for filename in os.listdir(hash_folder):
    if filename.endswith('.jpg'):
        image_path = os.path.join(hash_folder, filename)

        # Calculate the MD5 hash for the image
        with open(image_path, 'rb') as file:
            image_data = file.read()
            md5_hash = hashlib.md5(image_data).hexdigest()

        # Read the second text file
        with open(second_file_path, 'r') as file:
            for line in file:
                line = line.strip()
                if md5_hash in line:
                    # Split the line into columns
                    columns = line.split('\t')

                    # Extract the desired values
                    species_id = columns[3]
                    bb_x1 = columns[4]
                    bb_y1 = columns[5]
                    bb_x2 = columns[6]
                    bb_y2 = columns[7]

                    # Create the content for the new text file
                    content = f"{species_id} {bb_x1} {bb_y1} {bb_x2} {bb_y2}"

                    # Create the new text file with the matching name
                    new_filename = os.path.splitext(filename)[0] + '.txt'
                    new_filepath = os.path.join(output_folder, new_filename)
                    with open(new_filepath, 'w') as new_file:
                        new_file.write(content)

                    break  # Found a match, no need to continue searching

print("Processing completed!")
```
